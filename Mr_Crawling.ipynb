```python
import requests as rq
from bs4 import BeautifulSoup
import lxmlitem_price

# 남자의 악세사리 잡화 사이트 
url = "http://www.mr-s.co.kr/product/list.html?cate_no=500"
url
```

```python
res = rq.get(url)
res.url
html = res.content
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
#%% 타이틀 가져오기 
soup_item = soup.find_all("div",class_='txt_info')
soup_item
```

```python
for items in soup_item:
    item_name = items.find("p",class_='name')
    print(item_name.text)
    item_price = items.find_all("li",class_=' xans-record-')
    len1 = len(item_price)
    if len1>=3:
        print(item_price[2].text)
    
```
